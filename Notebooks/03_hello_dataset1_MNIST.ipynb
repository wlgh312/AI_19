```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
```

* 다운로드받기

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)#하나만 값이 true이고 나머지는 false
```

```python
print(mnist.train.images.shape)
print(mnist.test.images.shape)
```

```python
print(mnist.train.labels.shape)#labels : 0~9까지의 index를 가진 확률의 그릇
print(mnist.test.labels.shape)
```

```python
idx = 888
```

```python
img1 = mnist.test.images[idx]
img1 = np.array(img1, dtype='float')
pixels = img1.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.title('mnist.test.images[{}]'.format(idx))
plt.show()
```

```python
label1 = mnist.test.labels[idx]
print(label1)
```

Students 
* one-hot encoding 은 무엇일까? 
* 아래 코드의 의미는?

```python
label2 = np.argmax(label1)
print(label2)
```

하나 더 연습하기

```python
np.argmax([0, 1, 6, 3, 2, -2, 5, 7, 10])#최대값의 인덱스
```

```python
idx=123
img1 = mnist.test.images[idx]
img1 = np.array(img1, dtype='float')
lbl1 = np.argmax(mnist.test.labels[idx])

pixels = img1.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
plt.title('mnist.test.images[{}] ---> {}'.format(idx, lbl1))
plt.show()
```
