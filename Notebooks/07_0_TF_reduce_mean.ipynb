### About reduce_mean 

The functionality of numpy.mean and tensorflow.reduce_mean are the *same*

```
import numpy as np
import tensorflow as tf
```

```
c = np.array([[3.,4], [5.,6], [6.,7]])
print(c)
```

```
m1 = np.mean(c, 1)
print(m1)
```

```
m2 = tf.reduce_mean(c,1)
with tf.Session() as sess:
    result = sess.run(m2)
    print(result)
```
