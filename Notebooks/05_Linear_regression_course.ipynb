```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

**X and Y data**

```
x_train = [1, 2, 3]

y_train = [2+0.1+3, 4-0.3+3, 6+0.15+3] #  Add some noise 
```

**Initialization**

```
#W = tf.Variable(tf.random_normal([1]), name='weight')rank가 1인 랜덤값 배열 생성
#b = tf.Variable(tf.random_normal([1]), name='bias')
w0 = -10.0;
b0 = 10.0;

W = tf.Variable(w0*tf.ones([1]), name='weight')
b = tf.Variable(b0*tf.ones([1]), name='bias')
```

**Our hypothesis XW+b**

```
hypothesis = x_train * W + b
```

**cost/loss function**

```
cost = tf.reduce_mean(tf.square(hypothesis - y_train))#cost function
```

**Optimizer**

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)#cost최소화
```

# 그래프구현코드
-----------------------------------------------------

# 그래프실행코드


**Launch the graph in a session**

```
sess = tf.Session()
```

**Initializes global variables in the graph.** 

```
sess.run(tf.global_variables_initializer())
```

```
vw=[]
vb=[]
for step in range(2001):
    sess.run(train)
    w1 = sess.run(W)[0] # slope
    b1 = sess.run(b)[0] # bias
    vw.append(w1)
    vb.append(b1)
    
    if step % 100 == 0:
        print(step, sess.run(cost), w1, b1)
```

**Complete training**

```
plt.plot(vw)
```

```
plt.plot(vb)
```

```
w1 = sess.run(W)[0] # slope
b1 = sess.run(b)[0] # bias
str1 = 'y = ' + str(w1) +'x + ' + str(b1)
print(w1, b1)
print(str1)
```

```
plt.figure(1)
plt.plot(x_train, y_train,'o')

x1 = np.linspace(np.min(x_train)-1, np.max(x_train)+1)
y1 = w1*x1 + b1
plt.plot(x1, y1)
plt.grid() 
plt.title(str1)
```
